### Problem 2: Geometry and fluxes of the four plumes

In this part of the exercise, you will determine the start, end, and center positions of the four plumes, the distance of their centers to the source, the integrated amount of C$_2$H$_2$ in each plume, and the angle $\alpha$ of each transect with respect to the main plume axis. 

**Task 2.1: Plume positions and distance from source**

The background values scatter around zero because the acetylene signal background was already subtracted beforehand for this exercise. Due to the noisy signal, extracting the plume transects is not straightforward. A simple option is to smooth the signals and determine the plumes as the coherent parts of the time series above a given threshold. Alternatively, you may identify the plumes "manually" in the data.

*Recipe for automatic plume detection:*
* Smooth C2H2_excess signal with a centered moving average (https://en.wikipedia.org/wiki/Moving_average) of width 21.
* Define the plumes as the parts of the signal above 0.17 ppb.

For each plume, determine the following quantities:
* Coordinates (in meters) of the start, end and center.
* Distance from the source (in meters) using Pythagoras' theorem

```python
width = 21
vmin = 0.17

# smoothed signal
data['C2H2_smooth']=data['C2H2_excess'].rolling(width,center=True).mean()
data['C2H2_smooth'].plot()

mask = data['C2H2_smooth'] > vmin
data['C2H2_smooth'][mask].plot(marker='.',linestyle='')
```

```python
# create a separate mask for each plume
plume1 = mask & (data.index < '2015-04-23 10:08')
plume2 = mask & (data.index > '2015-04-23 10:08') & (data.index < '2015-04-23 10:10')
plume3 = mask & (data.index > '2015-04-23 10:12') & (data.index < '2015-04-23 10:14')
plume4 = mask & (data.index > '2015-04-23 10:15') & (data.index < '2015-04-23 10:17')
plumes = [plume1,plume2,plume3,plume4]

# overlay plumes on original (non-smoothed) data
data['C2H2_excess'].plot()
data['C2H2_excess'][plume1].plot(marker=".",linestyle='')
data['C2H2_excess'][plume2].plot(marker=".",linestyle='')
data['C2H2_excess'][plume3].plot(marker=".",linestyle='')
data['C2H2_excess'][plume4].plot(marker=".",linestyle='')
```

```python
# compute coordinates of the plumes and their distances from the source
plume_list = []

for pl in plumes:
    pls=data.index[pl].min()
    ple=data.index[pl].max()
    plxs = data['x'][data.index == pls].values
    plys = data['y'][data.index == pls].values
    plxe = data['x'][data.index == ple].values
    plye = data['y'][data.index == ple].values
    plxc = (plxs+plxe)/2.
    plyc = (plys+plye)/2.
    print("Plume from %s to %s" % (pls,ple))
    print("Start coordinate  (x,y) = (%i,%i)" % (plxs, plys))
    print("End coordinate    (x,y) = (%i,%i)" % (plxe, plye))
    print("Center coordinate (x,y) = (%i,%i)" % (plxc, plyc))
    print("Distance from source    = %i m" % math.sqrt(plxc**2 + plyc**2))
    print()
    plentry = {'start':pls,'end':ple,'xs':plxs,'ys':plys,'xe':plxe,'ye':plye,
              'xc':plxc,'yc':plyc,'dist':math.sqrt(plxc**2 + plyc**2),
               'angle':0.,'scale':0.,'mask':pl}
    plume_list.append(plentry)
```

**Task 2.2: Angle $\alpha$ between street and plume axis**

Determine the angle $\alpha$ between the street and the axis of the plume (see illustration below).

Question: 

* Why are we interested in this angle?

The angle $\alpha$ can best be calculated as the difference between the angles $\alpha_2$ and $\alpha_1$, which are both angles of a right-angled triangle (https://en.wikipedia.org/wiki/Right_triangle) with one side aligned with our coordinate system. The angles can be computed from the coordinates of the plume center and from the difference between start and end positions of the plume, respectively.

*Geometry of street with respect to plume axis. The red point is the location of the source, the blue point the center of the plume.*




```python
# computation of alpha for each plume
for pl in plume_list:
    alpha2 = math.degrees(math.atan(pl['yc']/pl['xc']))
    alpha1 = math.degrees(math.atan((pl['ye']-pl['ys'])/(pl['xe']-pl['xs'])))
    pl['alpha'] = alpha2-alpha1
    print("Plume starting at %s: alpha = %f" % (pl['start'],pl['alpha']))
```

**Task 2.3: Plume integrals and fluxes**

The measurements from the car are taken at a fixed altitude above surface and thus transect the plume along a single line. Because we cannot sample the vertical extent of the plume, we cannot compute the total flux of C$_2$H$_2$ through the grey vertical plane in the sketch shown below, but we can compute the flux through a control surface with a height of 1 m and a horizontal axis spanning the width of the plume along the car transect.

The flux is then given as the product of the mass of C$_2$H$_2$ per meter distance perpendicular to the control surface (i.e. the mass within the grey block) times the wind speed perpendicular to the plain. Note that this speed is smaller than the measured wind speed if the wind direction is not perfectly perpendicular to the street, i.e. if $\alpha$ differs from 90°. 



For each of the four plumes, determine the following quantities:
* Mean measured wind speed $ws$
* Integral $I$ (in units of $\mu$g/m) of C2H2_excess along the transect, i.e. $\sum{(c_i[C_2H_2]\cdot \Delta{x_i}} \cdot \Delta{z})$, where $c_i$ is the concentration of point $i$ in units of $\mu$g/m$^3$, $\Delta{x_i}$ is the distance driven by the car at this point, and $\Delta{z}$=1 m is the vertical extent of the control surface.
* Geometric factor $s$ to convert the measured wind speed into its component perpendicular to the control surface.
* The flux through the control surface in units of $\mu$g/s given by $I \cdot ws \cdot s$

```python
# Computation of integrals (ug/m), mean wind speeds (m/s), 
# scaling factors, and finally C2H2 mass fluxes for each plume

# scaling factor to convert ppb to ug/m3 assuming STP conditions
MC2H2= 26.04
Md = 28.966
R = 8.3145
ref_t = 273.15
ref_p = 101325.00
ppb2ugm3 = ref_p / (R*ref_t) * MC2H2 / 1e3

# loop over the plumes
for pl in plume_list:
    # geometric factor
    pl['scale'] = math.fabs(math.sin(math.radians(pl['alpha'])))
    # integrate measured C2H2 along transect
    integral = 0.
    for idx, rows in data[pl['mask']].iterrows():
        loc = data.index.get_loc(idx)
        # compute distance driven by central differencing
        dist = math.fabs(data.iloc[loc-1]['x']-data.iloc[loc+1]['x'])/2.
        integral = integral + dist * data.iloc[loc]['C2H2_excess']
    pl['integral'] = integral * ppb2ugm3
    pl['ws'] = data[pl['mask']]['ws'].mean()
    pl['flux'] = pl['integral'] * pl['ws'] * pl['scale']
    print("Plume at %s: integral = %f, scale = %f, wind speed = %f" % 
          (pl['start'],pl['integral'],pl['scale'],pl['ws']))
    
print()
for pl in plume_list:
    print("Plume at %s: flux = %f ug/s" % (pl['start'],pl['flux']))
```

### Problem 3: Estimating the source strength using a Gaussian plume model

In this part of the exercise, we apply a Gaussian plume model to simulate the observations and estimate the emission strengths.

The Gaussian plume solution is based on the following assumption:
* Homogeneous flow in x direction
* Constant emissions located at a single point in space
* No sources and sinks during transport
* Stationary conditions
* Homogeneous turbulence $K_c$ in y and z direction. Neglected for x.
* Molecular diffusion negligible

The equation of the Gaussian plume we will apply in this exercise and an illustration for the situation of a plume emanating from a stack are presented below. In our case, we assume the position of the source to be ($x_s$,$y_s$,$z_s$)=(0,0,2), i.e. we assume the source to be in the center of our coordinate system and at 2 m above surface.

In the equation of the Gaussian plume, the x-axis corresponds to the mean wind direction (plume centerline). The plume expands perpendicular to the centerline into the horizontal (y-axis) and vertical (z-axis) direction. The horizontal and vertical spread of the plume at distance $x$ is given by the parameters $\sigma_y(x)$ and $\sigma_z(x)$, which are both functions of $x$ based on empirically derived formula. The equation has two exponential terms for the vertical direction $z$. The second of these terms represents the reflection of the plume at the surface, which is particularly important for a source near the ground (as in our case).

To compare the simulation with the mobile measurements, we need to sample the Gaussian plume at the distance $x$ between source and plume center and at the same altitude as the car measurements (assumed to be $z$ = 1.5 m) along the y-axis. By setting $\overline{u} = ws$, i.e. the mean wind speed measured during the plume transect, we can then calculate the flux through a control surface extending in z-direction over the range [1,2] meters and in y-direction over $[-4\sigma_y,4\sigma_y]$ m. The flux through this surface should then be comparable to the flux calculated for the measured plume.






**Task 3.1: Plot a cross-section of a Gaussian plume**

Compute and plot a Gaussian plume in a vertical plane perpendicular to the main plume axis. 

We make the following assumptions:

* Distance from the source $x$=150 m
* $\sigma_y$ and $\sigma_z$ can be parameterized using a formulation by Briggs (1973) (see Equation below).
* Neutral boundary layer (Pasquill class D). For this class, Briggs (1973) suggest the following parameterizations:
<br>$\sigma_y = \frac{0.08 x}{\sqrt{1+0.0001x}}$, $\sigma_z = \frac{0.06 x}{\sqrt{1+0.0015x}}$
* Source strength $Q_s$ = 0.1 g/s, which corresponds to 360 g/h
* Compute Gaussian plume values for a regular grid spanning the vertical plane at regular 1 m intervals with
$y$ in range [ -4$\sigma{_y}$, 4$\sigma{_y}$] and z in range [0,4$\sigma{_z}$]


*General form of parameterization of plume width parameters according to Briggs (1973)*




```python
import numpy as np

def gaussian_plume(x,ypos,zpos,u=1.,zs=2.0,Qs=0.1,alpha_y=0.08,beta_y=0.0001,
                   alpha_z=0.06,beta_z=0.0015,gamma_z=-0.5):
    
    # Gaussian plume using Briggs formula for width parameters
    sigma_y = alpha_y * x / math.sqrt(1.+beta_y*x)
    sigma_z = alpha_z * x * (1.+beta_z*x)**gamma_z
    
    # create array of points spanned by ypos and zpos
    gaussian = np.zeros((len(zpos),len(ypos)))
    
    # convert source from kg/h to g/s
    term1 = Qs/(2*math.pi*sigma_y*sigma_z*u)
    denom_y = 2*sigma_y**2
    denom_z = 2*sigma_z**2
    for iy,y in np.ndenumerate(ypos):
        for iz,z in np.ndenumerate(zpos):
            gaussian[iz,iy] = term1 * math.exp(-y**2/denom_y) * (
                math.exp(-(z-zs)**2/denom_z) + math.exp(-(z+zs)**2/denom_z))

    # return valus in units of ug/m3
    return gaussian*1e6

# define parameters
x = 150.   # distance from source (m)
zs = 2.    # altitude of release above surface (m)
u = 1.0    # wind speed m/s
Qs = 0.1   # source strength g/s

# Briggs parameters for neutral boundary layer (class D)
alpha_y = 0.08
beta_y = 0.0001
alpha_z = 0.06
beta_z = 0.0015
gamma_z = -0.5

# define plume points spread over distance [-4 sigma_y,4 sigma_y]
# in horizontal direction and [0,zs+4 sigma_z] in vertical direction
sigma_y = alpha_y * x / math.sqrt(1.+beta_y*x)
sigma_z = alpha_z * x * (1.+beta_z*x)**gamma_z
ypos = np.arange(np.floor(-4* sigma_y),np.ceil(4*sigma_y))+0.5
zpos = np.arange(0,4*sigma_z+zs)+0.5

# compute Gaussian
gaussian = gaussian_plume(x,ypos,zpos,u=u,zs=zs,Qs=Qs,alpha_y=alpha_y,
                          beta_y=beta_y,alpha_z=alpha_z,beta_z=beta_z,
                          gamma_z=gamma_z)

# plot Gaussian
fig, ax = plt.subplots(figsize=(10, 6))

levels = np.arange(0, 340, 10)
h = ax.contourf(ypos,zpos,gaussian,levels=levels)
ax.set_xlabel('y (m)')
ax.set_ylabel('z (m)')
cbar = plt.colorbar(h)
cbar.set_label('C$_2$H$_2$ ($\mu$g/m$^3$)')
```

* Repeat plume simulation for a neutral boundary layer. How does this affect the width of the plume?

```python
# Briggs parameters for unstable boundary layer (class A)
alpha_y = 0.22
beta_y = 0.0001
alpha_z = 0.2
beta_z = 0.
gamma_z = 1.
sigma_y = alpha_y * x / math.sqrt(1.+beta_y*x)
sigma_z = alpha_z * x * (1.+beta_z*x)**gamma_z
ypos = np.arange(np.floor(-4* sigma_y),np.ceil(4*sigma_y))+0.5
zpos = np.arange(0,4*sigma_z+zs)+0.5

# re-compute and plot Gaussian
gaussian = gaussian_plume(x,ypos,zpos,u=u,zs=zs,Qs=Qs,alpha_y=alpha_y,
                          beta_y=beta_y,alpha_z=alpha_z,beta_z=beta_z,
                          gamma_z=gamma_z)
fig, ax = plt.subplots(figsize=(10, 6))

levels = np.arange(0, 34, 1)
h = ax.contourf(ypos,zpos,gaussian,levels=levels)
ax.set_xlabel('y (m)')
ax.set_ylabel('z (m)')
cbar = plt.colorbar(h)
cbar.set_label('C$_2$H$_2$ ($\mu$g/m$^3$)')
```

```python
# flux through control surface for above plume
control_flux = gaussian[1,:].sum()*u
print('Flux through control flux centered at %f m = %f ug/s' % (zpos[1], control_flux))
```

```python
fluxes = []

for pl in plume_list:
    # define parameters
    x = pl['dist']
    u = pl['ws']
    sigma_y = alpha_y * x / math.sqrt(1.+beta_y*x)
    sigma_z = alpha_z * x * (1.+beta_z*x)**gamma_z
    ypos = np.arange(np.floor(-4* sigma_y),np.ceil(4*sigma_y))+0.5
    zpos = np.arange(0,4*sigma_z+zs)+0.5
    
    # compute Gaussian
    gaussian = gaussian_plume(x,ypos,zpos,u=u,zs=zs,Qs=Qs,alpha_y=alpha_y,
                              beta_y=beta_y,alpha_z=alpha_z,beta_z=beta_z,
                              gamma_z=gamma_z)
    sim_flux = gaussian[1,:].sum()*u
    fluxes.append({'sim_flux':sim_flux,'obs_flux':pl['flux'],
                   'source':Qs*pl['flux']/sim_flux*3600.})
    
print('Fluxes estimated for the four plumes:')
for f in fluxes: print("%8.2f ug/s" % f['sim_flux'])
print()
print('Source strength estimated for the four plumes:')
for f in fluxes: print("%8.2f g/h" % f['source'])
print()
print('Our final answer is the mean:')
smean = 0.
for f in fluxes: smean = smean + f['source']
print("%8.2f g/h" % (smean/4.))
```
